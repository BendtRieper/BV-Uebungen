1. Bild laden und Ausschnitt speichern

```python
import numpy as np
from skimage.io import imread, imsave
import matplotlib.pyplot as plt

img = imread("astronaut.png")
plt.imshow(img, cmap="gray", vmin=0, vmax=255)

abzeichen =  img[340:440, 120:220]
imsave('abzeichen.png', abzeichen)
```

Ausschnitt anzeigen

```python
plt.imshow(abzeichen, cmap="gray", vmin=0, vmax=255)
```

2. Bild manipulieren

```python
astronaut_kopie = np.copy(img)

astronaut_kopie[100,100] = 0

plt.imshow(astronaut_kopie, cmap="gray", vmin=0, vmax=255)
```

```python
astronaut_kopie_helm = np.copy(img)

astronaut_kopie[320:,270:] = 0

plt.imshow(astronaut_kopie, cmap="gray", vmin=0, vmax=255)
```

3. Neues Bild nach Vorgabe erstellen


a) vertikal spiegeln

```python
spiegelung_vertikal = np.flip(img,1)
plt.imshow(spiegelung_vertikal, cmap="gray", vmin=0, vmax=255)
imsave('spiegelung_vertikal.png', spiegelung_vertikal)
```

b) Spiegelung Horizontal

```python
spiegelung_horizontal = np.flip(img,0)
plt.imshow(spiegelung_horizontal, cmap="gray", vmin=0, vmax=255)
imsave('spiegelung_horizontal.png', spiegelung_horizontal)
```

c) Nacheinanderausführung

```python
spiegelung_beides = np.flip(img,1)
spiegelung_beides = np.flip(spiegelung_beides,0)
plt.imshow(spiegelung_beides, cmap="gray", vmin=0, vmax=255)
imsave('spiegelung_beides.png', spiegelung_beides)
```

d) neues Bild erzeugen

```python
new_img =  np.zeros((img.shape[0]*2, img.shape[1]*2), dtype=img.dtype)

new_img[:img.shape[0], :img.shape[1]] = spiegelung_horizontal #links oben
new_img[:img.shape[0], img.shape[1]:] = spiegelung_beides #rechts oben
new_img[img.shape[0]:, :img.shape[1]] = img #unten links
new_img[img.shape[0]:, img.shape[1]:] = spiegelung_vertikal #unten rechts

plt.imshow(new_img, cmap="gray", vmin=0, vmax=255)
```
