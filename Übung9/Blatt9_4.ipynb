```python
import numpy as np
import matplotlib.pyplot as plt
from math import sin,cos, pi

#Der Plot geht von + 4 bis -2 also hat er eine Amplitude von 3 und eine Verschiebung nach oben um 1 -> wir addieren 1 auf die Funktion
#Ein Phasendurchlauf ist ein bisschen mehr als 4 Pi, also haben wir ein Sinusoid mit einer Frequenz von weniger als 0.5 Pi
#Dieser Sinusoid hat eine Amplitude von 2, da die "Mittellinie" von 3 bis -1 geht
#Der naechste Sinusoid hat eine hoehere Frequenz
#Wir brauchen eine Phasenverschiebung, da der Graph eine Verschiebung auf der y-Achse hat, jedoch trotzdem durch 0,0 geht


def sinusoide_fkt(x):
  return sin(3*x - pi/2) + 2*sin(0.4*x) + 1

x = np.linspace(0, 10*pi, 1000)
y_norm = list(map(sinusoide_fkt, x))

plt.plot(x, y_norm)
plt.grid()
plt.ylim(-5,5)


```
