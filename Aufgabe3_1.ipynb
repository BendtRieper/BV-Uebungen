```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imsave

#Importieren von den bentigten Bildern
# 1   0,450,52 m, Blau-Grn
# 2   0,520,60 m, Grn
# 3   0,630,69 m, Rot
# 4   0,760,90 m, nahes Infrarot
# 5   1,551,75 m, mittleres Infrarot
# 6   10,412,5 m Thermal
# 7   2,082,35 m, mittleres Infrarot

rot_1 = imread('/content/bands1/band3.png')
nir_1 = imread('/content/bands1/band4.png')
swir_1 = imread('/content/bands1/band7.png')


ndvi_1 = (nir_1 - rot_1) / (nir_1 + rot_1)

ndwi_1 = (nir_1 - swir_1) / (nir_1 + swir_1)

#ndvi Bild anzeigen
plt.imshow(ndvi_1, cmap="gray", vmin=-1, vmax=1)
ndvi_1
```

```python
#ndwi Bild Anzeigen
plt.imshow(ndwi_1, cmap="gray", vmin=-1, vmax=1)
```
