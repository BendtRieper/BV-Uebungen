1. Hintergund "erstellen"

```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imsave

# Laden der Bilder

bild1=imread('bild1.png')
bild2=imread('bild2.png')
bild3=imread('bild3.png')
bild4=imread('bild4.png')

# Anwendung Gesetz der großen Zahlen

mittelwert = (bild1+bild2+bild3+bild4)/4

plt.imshow(mittelwert, cmap="gray", vmin=0, vmax=255)
```

2. Besseren Hintergund mit Median erzeugen

```python
medianArray = np.zeros((bild1.shape[0],bild1.shape[1],4))

medianArray[:,:,0] = bild1
medianArray[:,:,1] = bild2
medianArray[:,:,2] = bild3
medianArray[:,:,3] = bild4

median = np.median (medianArray, axis=-1)

plt.imshow(median, cmap="gray", vmin=0, vmax=255)
```

3. Veränderungen ermitteln

```python
# hoher threshold = größere Veränderungen
# niedrieger threshold = erkennen von kleinen Veränderungen
aebild1 = np.abs(bild1 - median) > 40
plt.imshow(aebild1, cmap="gray", vmin=0, vmax=1)
```

```python
aebild2 = bild2 - median > 40
plt.imshow(aebild2, cmap="gray", vmin=0, vmax=1)
```

```python
aebild3 = bild3 - median > 30
plt.imshow(aebild3, cmap="gray", vmin=0, vmax=1)
```

```python
aebild4 = bild4 - median > 10
plt.imshow(aebild4, cmap="gray", vmin=0, vmax=1)
```

4. Hintergund Bild "zusammenbauen"

```python
hintergrund_copy = median.copy()

hintergrund_copy[aebild1]=bild1[aebild1]
hintergrund_copy[aebild2]=bild2[aebild2]
hintergrund_copy[aebild3]=bild3[aebild3]
hintergrund_copy[aebild4]=bild4[aebild4]
plt.imshow(hintergrund_copy, cmap="gray", vmin=0, vmax=255)
```
