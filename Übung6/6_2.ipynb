2. Histogrammausgleich

```python
def hist_equ (img):
    img = (255 * img).astype(np.uint8)

    # Compute the histogram and CDF of the image
    hist, _ = np.histogram(img, bins=256, range=(0, 255))
    cdf = hist.cumsum()
    cdf = cdf / cdf[-1]  # Normalize the CDF

    # Compute the transformation function
    T = (255 * cdf).astype(np.uint8)

    # Apply the transformation function to the image
    equalized_img = T[img]

    return equalized_img, cdf, T

equalized_bild1, cdf, T1 = hist_equ(bild1)
equalized_bild3, cdf, T3 = hist_equ(bild3)

plt.imshow(equalized_bild1, cmap='gray')
plt.show()

# Plotte die CDF
plt.title('Kumulative Verteilungsfunktion Bild 1')
plt.plot(cdf)
plt.show()

plt.imshow(equalized_bild3, cmap='gray')
plt.show()

# Plotte die CDF
plt.title('Kumulative Verteilungsfunktion Bild 3')
plt.plot(cdf)
plt.show()

plt.title('Kumulative Verteilungsfunktion Bild 3')
plt.plot(T1)
plt.plot(T3)

```

Man sieht sehr gut, das wir durch den Histogrammausgleich ein besseres ergebnis erzielen. Da Gebäude wird deutlich besser aufbereitet und das ganze Bild ist nicht so grau (einfarbig ) wie in der Lösung. Auch der Hubscharuber ist besser zu erkennen. Die Pflanzen sind deutlich besser zu erkennen, der Hubscharuber auch. Hier sind mehr Konturen zu sehen.
