```python
import numpy as np
from skimage.io import imread, imsave
import matplotlib.pyplot as plt

img = imread("astronaut.png")
plt.imshow(img, cmap="gray", vmin=0, vmax=255)
```

```python
# es wird das Bild übergen und die Standardabweichung der Normalverteilung
def gaussian(image, sigma):
 noise = np.random.normal(0, sigma, (img.shape[0],img.shape[1])) 
 return (image + noise)


astro_gaus = gaussian(np.copy(img), 30)
#erzwingt die Werte auf Graustufen von 0 bis 255
np.clip(astro_gaus, 0, 255)
plt.imshow(astro_gaus, cmap="gray", vmin=0, vmax=255)
```

Je höher der Wert der Standardabweichung, desto flacher und breiter ist der Graph der Normalverteilung. Es können somit Werte gewählt werden, die mehr vom Mittelwert abweichen.

```python
import random

# es wird das Bild übergeben und die Prozentzahl der Anzahl der Pixel die manipuliert werden sollen von 0 bis 1
def add_noise(img, p):
    number_rows = img.shape[0]
    number_cols = img.shape[1]

    pixels_in_image = img.shape[0] * img.shape[1]

    # Anzahl Pixel berechnen die manipuliert werden sollen 
    number_of_pixels = p * pixels_in_image
    n_o_p = int(round(number_of_pixels))

    #Hälfte der Pixel sollen weiß gefärbt werden
    for i in range(int(n_o_p/2)):
        
        y_coord=random.randint(0, number_rows - 1)
        x_coord=random.randint(0, number_cols - 1)
        img[y_coord][x_coord] = 255

     #Hälfte der Pixel sollen schwarz gefärbt werden     
    for i in range(int(n_o_p/2)):
        
        y_coord=random.randint(0, number_rows - 1)
        x_coord=random.randint(0, number_cols - 1)
        img[y_coord][x_coord] = 0
          
    return img

astro_sp = add_noise(np.copy(img), 0.1)
plt.imshow(astro_sp, cmap="gray", vmin=0, vmax=255)


```

Die Wahrscheinlichkeit für eine Veränderung bestimmt die Anzahl der Pixel welche vom Rauschen betroffen sind. Umso höher die Wahrscheinlichkeit für das Bild ist, desto mehr Pixel sind betroffen.
