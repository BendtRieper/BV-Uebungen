```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.io import imread, imsave

img1 = imread('bildverbesserung/bild1.png')   #f(x) = (x - min_val) * (new_max - new_min) / (max_val - min_val) + new_min
min_val = np.min(img1)
max_val = np.max(img1)
new_max = 1.0
new_min = 0.0
stretched = (img1 - min_val) *  (new_max - new_min) / (max_val - min_val)+0
plt.imshow(stretched, cmap="gray", vmin=new_min, vmax=0.1)



```

2. Thresholding

```python
img2 = imread('bildverbesserung/bild2.png')

maske = (img2>60) & (img2<110)
img2[maske] = 0

plt.imshow(img2, cmap="gray", vmin=0, vmax=255)
```
